```python
# Deleting the unnecessary columns
df= df[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude',
                 'AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]

df.head()
```

# Exploratory Data Analysis

## 1. Terrorist Activities According to the Year

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',stacked=False, figsize=(15,6))
plt.title('Terrorist Activities', Fontsize=25)
plt.ylabel('Number of Attacks', Fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.xticks([1970,1975,1980,1985,1990,1995,2000,2005,2010,2015])
ax=plt.axes()
ax.set_facecolor('#f2fafd')
plt.show()
```

```python
year = df['Year'].unique()
years_count = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,7))
sns.barplot(x = year, y = years_count, palette = "viridis")
plt.xlabel('Attacking Year',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.title('Terrorist Activities According to the Year',fontsize=25)
plt.xticks(rotation = 50)
plt.show()
```

Number of attacks were **HIGHEST** in the year **2014**

## 2. Most Affected Region

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(df['Region'].value_counts()[:10].index,df['Region'].value_counts()[:10].values, palette="mako_r")
plt.title('Top 10 Affected Regions', Fontsize=25)
plt.xlabel('Region', Fontsize=15)
plt.ylabel('Count', Fontsize=15)

#Rotate the labels on X-axis so that they don't overlap
plt.xticks(rotation=90);
```

MOST affected regions are **Middle East & North Africa**

## 3. Top 10 Affected Countries 

```python
#Gets the individual count of each country
country_count =  df['Country'].value_counts()

#Get the top 10 countries by sclicing
country_count10 = country_count[:10]
print(country_count10)
```

```python
# Create Figure (empty canvas)
fig = plt.figure(figsize=(10,6))
sns.barplot(df['Country'].value_counts()[:10].index,df['Country'].value_counts()[:10].values, palette="magma")
plt.title('Top 10 Countries Affected', Fontsize=25)
plt.xlabel('Countries', Fontsize=15)
plt.ylabel('Count', Fontsize=15)

#Rotate the labels on X-axis so that they don't overlap
plt.xticks(rotation=90);
```

**Iraq** is the MOST affected country

## 4. Most Affected States 

```python
print('Most Affected States are:\n', df['State'].value_counts().head(10))
```

```python
# Create Figure (empty canvas)
fig = plt.figure(figsize=(15,6))
sns.barplot(df['State'].value_counts()[:10].index,df['State'].value_counts()[:10].values,palette='rocket_r')
plt.title('Top 10 States Affected', Fontsize=25)
plt.xlabel('States', Fontsize=15)
plt.ylabel('Count', Fontsize=15)

#Rotate the labels on X-axis so that they don't overlap
plt.xticks(rotation=90);
```

**Baghdad** is the MOST affected state due to Terrorism.

## 5. Methods of Attack

```python

fig = plt.figure(figsize=(15,6))
sns.countplot('AttackType',data=df,order=df['AttackType'].value_counts().index,palette='vlag')
plt.title("Most Used Attack Method", Fontsize=25)
plt.xlabel('Methods', Fontsize=15)
plt.ylabel('Count', Fontsize=15)

#Rotate the labels on X-axis so that they don't overlap
plt.xticks(rotation=90)
```

**Bombing/Explosion** are MOSTLY used as an attacking method

## 6. Types of Target

```python
fig = plt.figure(figsize=(15,6))
sns.countplot('Target_type',data=df,order=df['Target_type'].value_counts().index,palette='coolwarm_r')
plt.title("Most Frequent Type of Target", Fontsize=25)
plt.xlabel('Target Type', Fontsize=15)
plt.ylabel('Count', Fontsize=15)

#Rotate the labels on X-axis so that they don't overlap
plt.xticks(rotation=90)
```

## 7. Most Frequent Groups involved in Terrorist Activity

```python
fig = plt.figure(figsize=(15,6))
sns.barplot(df['Group'].value_counts()[:10].index,df['Group'].value_counts()[:10].values,palette='rocket')
plt.title('Most Active Terrorism Groups', Fontsize=25)
plt.xlabel('Groups', Fontsize=15)
plt.ylabel('Count', Fontsize=15)

#Rotate the labels on X-axis so that they don't overlap
plt.xticks(rotation=90);
```
